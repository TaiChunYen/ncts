### 四種屬性寶可夢hp,speed分佈圖



```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = pd.read_csv('./Pokemon.csv')
```

```python
gx=data[data['Type 1']=='Grass'].HP
gy=data[data['Type 1']=='Grass'].Speed
```

```python
fx=data[data['Type 1']=='Fire'].HP
fy=data[data['Type 1']=='Fire'].Speed
```

```python
wx=data[data['Type 1']=='Water'].HP
wy=data[data['Type 1']=='Water'].Speed
```

```python
ex=data[data['Type 1']=='Electric'].HP
ey=data[data['Type 1']=='Electric'].Speed
```

```python
plt.scatter(gx,gy,c='g',label='gress')
plt.scatter(wx,wy,c='b',label='water')
plt.scatter(fx,fy,c='r',label='fire')
plt.scatter(ex,ey,c='y',label='electric')
plt.xlabel('HP')
plt.ylabel('Speed')
plt.legend()
```
